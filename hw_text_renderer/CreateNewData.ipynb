```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
path = "ETL1/data/"
files_images = os.listdir("ETL1/data/images")
files_labels = os.listdir("ETL1/data/labels")
```

```python
images = []
for image in files_images:
   img = cv2.imread(path+"images/"+image)
   images.append(img)
```

```python
print(images[0])
```

```python
plt.imshow(images[0])
```

```python
labels = []
for file in files_labels:
    with open(path + "labels/" + file,"r") as f:
        labels.append(f.read())

print(labels[-1])
```

```python
from sklearn.utils import shuffle
images, labels = shuffle(images, labels)
```

```python
def imshow(title, img, size =10):
    w,h = img.shape[0], img.shape[1]
    tlkh = w/h
    plt.figure(figsize=(tlkh * size, size))
    plt.title(title)
    plt.imshow(img)

for i in range(20):
    imshow(labels[i], images[i])
```

```python
# tạo ảnh có 5 số
new_image = np.zeros([images[0].shape[0], images[0].shape[1] * 5, images[0].shape[2]])
label_image = []
for i in range(len(images)):
    if i+4 == len(images):
        break
    new_image[:,:images[0].shape[1]] = images[i]
    new_image[:,images[0].shape[1]:images[0].shape[1]*2] = images[i+1]
    new_image[:,images[0].shape[1]*2:images[0].shape[1]*3] = images[i+2]
    new_image[:,images[0].shape[1]*3:images[0].shape[1]*4] = images[i+3]
    new_image[:,images[0].shape[1]*4:images[0].shape[1]*5] = images[i+4]


    if i % 5 == 0:
        cv2.imwrite("MyData/"+str(i)+"-5.png",new_image)
        f = open("newLabels/{}-5.txt".format(i),"w")
        f.writelines( "MyData/"+str(i)+"-5.png" + ", " +""""{}{}{}{}{}" """.format(labels[i], labels[i+1], labels[i+2], labels[i+3], labels[i+4]))
        print("save img",str(i)+"-5")
        new_image = np.zeros([images[0].shape[0], images[0].shape[1] * 5, images[0].shape[2]])
```
